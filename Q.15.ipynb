```python
 # Write a function to check if a given year is a leap year.  
# A leap year is divisible by 4, but not by 100, unless also divisible by 400.  
# Divisible by 4: If a year can be evenly divided by 4 (like 2020), it might be a leap year.  
# Not Divisible by 100: However, if that year can also be evenly divided by 100 (like 1900),
# then it's not a leap year.
```

```python
year = 2016
if year>= 1582 and (year%4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(True)
else:
    print(False)
```

```python
def is_leap_year(year:int) -> bool:
    """
    This fucntion checks year in leep year.
    """
    if not isinstance(year, int):
        raise TypeError("Year must be a integer value")
    if year < 1582:
        raise ValueError("Year must be integer of 1582")
    
    if year>= 1582 and (year%4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

if __name__ == "__main__":
    user_input = 2012
    try:
        result = is_leap_year(user_input)
        if result:
            print(f"{user_input} is a leep year")
        else:
            print(f"{user_input} is not a leep year")
    except (TypeError, ValueError) as error:
        print("Error", error)
    except Exception as e:
        print(e)
        
        
```
