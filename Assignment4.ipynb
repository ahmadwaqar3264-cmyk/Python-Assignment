# 1. Positional Arguments

```python
def function(a, b, c):
    print(a + b + c)
```

```python
function(2,5,6)
```

# 2. Keyword Arguments

```python
def student(name, age):
    print(name, age)
```

```python
student(age = 13, name = "Sahil")
```

# 3. Positional only Arguments ( / )

```python
def function(a, b, /):
    print(a+b)
```

```python
function(2,4)
```

# 4. Keyword only Arguments ( * )

```python
def function(*, a, b):
    print(a+b)
```

```python
function(a= 1, b= 13)
```

# 5. Default Arguments

```python
def info(name, city="Delhi"):
    print(name,city)
```

```python
info("rohit")
info("rohit","Banglore")
```

# 6. Arbitrary Arguments (* Args)

```python
def total(*number):
    print(sum(number))
```

```python
total(1,3,4,5,6)
```

# 7. Arbitrary Keyword Arguments (** kwargs)

```python
def employee(**kwargs):
    print(kwargs)
```

```python
employee(name="Waqar", age=25, city="Mau")
```

# Q.1. Find the largest number in the List?

```python
list = [88,1,3,4,5,6,7,8,9,3,4,5,6,7,99,8,9,33,6,78,9,56,36,99]

largest = list[0]

for item in list:
    if largest < item:
        largest = item

print("largest number is:", largest)   
```

# Q.2. Find Even number from a List

```python
list1 = [1,3,4,5,6,7,8,9,3,4,5,6,7,8,9]
even_list = []
for item in list1:
    if item % 2 == 0:
        even_list.append(item)
print("even list is:",even_list)        
```

# Q.3. Find out all string values from a List?

```python
list1 = ["Python",2,4,4.5,7,8,67,[1,2,3],"Mysql"]
for item in list1:
    if type(item) == str:
        print(item)
```

# Q.4. Fliter Unique Element from a List?

```python
list1 = [1,3,4,5,6,7,8,9,3,4,5,6,7,8,9]
unique_list = []
for item in list1:
    if item in unique_list:
        continue
    else:
        unique_list.append(item)
print(unique_list)        
```

# Q.5. Flatten a Nested List?

```python
listw = [[9, 2, 4],[4, 44, 989],[6, 7, 4567]]
```

```python
flatten_list = []
for sublist in listw:
    for item in sublist:
        flatten_list.append(item)
print(flatten_list)        
    
```
