```python
def find_largest_numbers(numbers_list: list[float]) -> float:
    """
    This function returns the largest number
    from a given list of numbers.
    """

    # check input is a list
    if not isinstance(numbers_list, list):
        raise TypeError("Input must be a list of numbers")

    # check list is not empty
    if not numbers_list:
        raise ValueError("List cannot be empty")

    # check all elements are numbers
    for num in numbers_list:
        if not isinstance(num, (int, float)):
            raise TypeError("All elements must be numeric")

    # find largest number
    largest = numbers_list[0]
    for num in numbers_list:
        if num > largest:
            largest = num

    return largest


if __name__ == "__main__":
    try:
        numbers_list = [1, 23, 12, 3, 4, 12, 12, 24, 43]

        result = find_largest_numbers(numbers_list)
        print("Largest number in the list is:", result)

    except (TypeError, ValueError) as error:
        print(error)

    except Exception as e:
        print("Unexpected error:", e)
```
