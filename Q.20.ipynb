```python
def gcd(a: int, b:int) -> int:
    """
    This function calculate GCD of two numbers.
    """
    if not isinstance(a,int) or not isinstance(b,int):
        raise TypeError("Input must be an integer")
    if a == 0 and b == 0:
        raise ValueError("GCD is not defined for both values as zeero")
    a, b = abs(a), abs(b)
    
    while b != 0:
        a,b = b,a%b
    return a
if __name__ == "__main__":
    a = 50
    b = 6
    try:
        result = gcd(a,b)
        print("GCD is :", result)
    except TypeError as e:
        print("Error",e)
    except ValueError as e:
        print("Error", e)
    except Exception as e:
        print("Somthing went wrong", e)
```
