```python
# Example: sum_of_digits(123) -> 6
```

```python
n = 123
total = 0
while n > 0:
    digit = n % 10
    total = total + digit
    n = n //10
print(total)
```

```python
def sum_of_digits(n: int) -> int:
    """
    This function calculate the sum of non-integer.
    """
    if not isinstance(n,int):
        raise TypeError("Input must be an integer")
    if n < 0:
        raise ValueError("Input must be a possititive number")
    total = 0
    while n > 0:
        digit = (n%10)
        total = total + digit
        n = n // 10
    return total  

if __name__ == "__main__":
    num = input("Enter your number : ")

    try:
        num = int(num)
    except ValueError:
        print("Error : " "please enter valid integer only.")
    else:
        try:
            result = sum_of_digits(num)
            print("Sum of digits", result)
        except (TypeError, ValueError) as error:
            print("Error", error)
        
```
