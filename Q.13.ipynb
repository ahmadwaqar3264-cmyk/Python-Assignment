```python
# Write a function to check if two input strings are anagrams of each other.  
# An anagram is a word or phrase that is formed by rearranging the letters of another word or phrase.
# To be anagrams, two strings must contain the exact same characters in the same quantities, but in
# any order.  Example: "listen" and "silent" are anagrams.
```

```python
string1 = "bca"
string2 = "acb"
char_count = {}
if len(string1) != len(string2):
    print(False)
else:
    for char in string1:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    print(char_count)
    for char in string2:
        if char not in char_count:
            print(False)
            break
        char_count[char] -= 1
        if char_count[char] < 0:
            print(False)
            break
    else:
        print(True)
        
```

```python
def is_anagram(string1: str, string2: str) -> bool:
    """
    This function check the string is anagram.
    """
    if not isinstance(string1, str) or not isinstance(string2, str):
        raise TypeError("Input must be string")
    string1 = string1.replace(" ", "").lower()
    string2 = string2.replace(" ", "").lower()

    if len(string1) != len(string2):
        return False
        
    char_count = {}

    for char in string1:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    print(char_count)
    for char in string2:
        if char not in char_count:
            return False
            break
        char_count[char] -= 1
        if char_count[char] < 0:
            return False
            break
    else:
        return True
        
if __name__ == "__main__":
    string1 = "care"
    string2 = "race"
    try:
        result = is_anagram(string1, string2)
        if result:
            print(result)
        else:
            print("Your string is not anagram")
    except TypeError as e:
        print(e)
    
        
```
