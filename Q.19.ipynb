```python
# Note: Don’t use split or join. 
# Write a function to reverse the order of words in a sentence, keeping the words themselves
# intact (sentence should not be changed or altered in any way; they should remain exactly as
# they are,).
# Example
# Input: "Python is fun"
# Expected Output: "fun is Python"
```

```python
sentence = "Python is fun"
words_list = []
word = ""
for char in sentence:
    if char != " ":
        word = word + char
    else:
        words_list.append(word)
        word = ""
words_list.append(word)        

rev_sentence = ""
for i in range(len(words_list)-1,-1,-1):
    rev_sentence = rev_sentence + words_list[i]
    if i != 0:
        rev_sentence = rev_sentence + " "
print(rev_sentence)    
```

```python
def reverse_word_function(sentence:str) -> str:
    """
    This function return reverse word.
    """
    if not isinstance(sentence,str):
        raise TypeError("Input must be a string")
    words_list = []
    word = ""
    for char in sentence:
        if char != " ":
            word = word + char
        else:
            words_list.append(word)
            word = ""
    words_list.append(word)        
    
    rev_sentence = ""
    for i in range(len(words_list)-1,-1,-1):
        
        rev_sentence = rev_sentence + words_list[i]
        if i != 0:
            rev_sentence = rev_sentence + " "
    return rev_sentence
if __name__ == "__main__":
    sentence = "python is Fun"
    try:
        result = reverse_word_function(sentence)
        print(result)
    except TypeError as e:
        print(e)
    except Exception as e:
        print("Error",e)
```
