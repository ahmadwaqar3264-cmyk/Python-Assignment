```python
number = 13
for i in range(2,int(number**.5)+1):
    if number % i == 0:
        print("Not a prime number")
        break
else:
    print("Prime number")
        
```

```python
def is_prime_number(number: int) -> bool:
    """
    Function check whether a number is prime or not.
    Args:
        number (int): The number to check.
    Returns:
         bool: True if the number is prime, False otherwise.
    Raises:
          TypeError: If input is not an integer.
          ValueError: If number is less than 2 (since primes are >= 2.)
          
    """
    if not isinstance(number, int):
        raise TypeError("Input must be an integer")
    if number < 2:
        raise ValueError("Prime numbers are greater than or equal to 2")
    for i in range(2, int(number**.5)+1):
        if number % i == 0:
            return False
            break
    else:
        return True
        
if __name__ == "__main__":
    try:
        user_input = input("Enter the number to check ites prime or not")
        if not user_input.strip().isdigit():
            raise TypeError("Input must be a digit")
        number = int(user_input)
        if is_prime_number(number):
            print(f"{number} is a prime number")
        else: 
            print(f"{number} is not a prime number")
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("Somthing went wrong",e)
            
        
```
