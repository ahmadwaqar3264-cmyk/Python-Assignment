```python
# Write a function that removes duplicates from a list while maintaining the original order.(Don’t use set method.)
```

```python
# Example
# Input: [1, 3, 2, 3, 4, 1, 5]
# Expected Output: [1, 3, 2, 4, 5]  Example
# Input: [4, 4, 4, 4]
# Expected Output: [4]
```

```python
list1 = [100, 2,3,4,4,5]
result = []
for i in list1:
    if i not in result:
        result.append(i)
print(result)        
```

```python
from typing import List,Any
def remove_duplicates(list1: List[Any]) -> List[Any]:
    """
    This function remove from list while maintaining orignal order.
    """
    if not isinstance(list1, list):
        raise TypeError("Input must be a list")
    if len(list1) == 0:
        raise ValueError("List cannot be empty")
    if len(list1) == 1:
        print("List has already one element. No duplicate to remove")
        return List
    result = []
    for i in list1:
        if i not in result:
            result.append(i)
    return result

if __name__ == "__main__":
    numbers_list = [100,3,2,2,4,5,7,10,10,10]
    try:
        result = remove_duplicates(numbers_list)
    except (TypeError, ValueError) as error:
        print("Error: ",error)
    except Exception as error:
        print("Somthing went wrong :", error)
    else:
        print("After removing duplicates : ", result)
        
```
