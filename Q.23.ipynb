```python
# Write a function that returns the intersection of two lists without using set operations.  Example:
# Input: list1 = [1, 2, 3, 4, 5], list2 = [4, 5, 6, 7, 8]
# Expected Output: [4, 5]
```

```python
list1 = [1, 2, 3, 4, 5,5]
list2 = [4, 5, 6, 7, 8]
result = []
for i in list1:
    if i in list2 and i not in result:
        result.append(i)
print(result)        
```

```python
def intersection(list1:list, list2:list) -> list:
    """
    This function returns common elements from both lists
    """
    if not isinstance(list1,list) or not isinstance(list2,list):
        raise TypeError("Both orguments must be lists")
    if not list1 or not list2:
        raise ValueError("Input must not be empty list")
    result = []
    for i in list1:
        if i in list2 and i not in result:
            result.append(i)
    return result
if __name__ == "__main__":
    list1 = [1, 2, 3, 4, 5,5]
    list2 = [4, 5, 6, 7, 8]
    try:
        result = intersection(list1,list2)
        print(result)
    except (TypeError, ValueError) as e:
        print("Error:",e)
    except Exception as error:
        print("Error:", error)
    
    
```
