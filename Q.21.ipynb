```python
# Write a function to find the Greatest Common Divisor (GCD) of two numbers using recursion.
# Example
# Input: a = 48, b = 18
# Expected Output: 6 (the GCD of 48 and 18 is 6)
```

```python
a = 48
b = 18
```

```python
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)
print(gcd(a,b))    
```
