```python
# Q7. Count Vowels in a String
# Take a string as input and count the number of vowels (a, e, i, o, u) in it.
# Ignore case (i.e., count both uppercase and lowercase vowels).
# Example: “Python” word has 1 vowel sound(o)
```

```python
text = "Python"
vowels = "aeiou"
count = 0
for char in text.lower():
    if char in vowels:
        count = count + 1
print(count)        
```

```python
def count_vowels(text: str) -> int:
    """
    This function take string to count vowels.
    """
    if not isinstance(text, str):
        raise TypeError("Input must be string")
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char in vowels:
            count = count + 1
    return count

if __name__ == "__main__":
    try:
        user_input = input("Enter your string to count vowels : ")
        result = count_vowels(user_input)
        print(f"total vowels", result)

    except TypeError as e:
        print("error", e)
    except Exception as e:
        print(e)
```
