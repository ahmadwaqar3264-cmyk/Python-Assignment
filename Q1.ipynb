```python
def reverse_maker(string1: str) -> str:
    """
    This function takes a string as input
    and return the reverse of the string
    using for loop.
         
    """
    # validate the input type
    if not isinstance(string1, str):
        raise TypeError("Input must be string")

    rev = ""
    for char in string1:
        rev = char + rev
    return rev

if __name__ == "__main__":
    user_input = "Large"
    try:
        return_value = reverse_maker(user_input)
        print(return_value)
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
    
```
