# Q.1

```python
units = int(input("Enter number of units: "))
bill = 0

if units > 200:
    bill = (units - 200) * 10 + 100 * 5
elif units > 100:
    bill = (units - 100) * 5 
else:
    bill = 0

print("Total bill amount: Rs", bill)
```

# Q.2

```python

percentage = float(input("Enter your percentage: "))

if percentage > 90:
    grade = "A"
elif percentage > 80 and percentage <= 90:
    grade = "B"
elif percentage >= 60 and percentage <= 80:
    grade = "C"
else:
    grade = "D"

print("Your grade is:", grade)
```

```python
salary = int(input("Enter salary: "))
years = int(input("Enter years of service: "))

if years > 10:
    bonus = salary * 10 / 100
elif years >= 6 and years <= 10:
    bonus = salary * 8 / 100
else:
    bonus = salary * 5 / 100

print("Net bonus amount is:", bonus)
```

# Q.6

```python
price = int(input("Enter marked price: "))

if price > 10000:
    discount = price * 20 / 100
elif price > 7000 and price <= 10000:
    discount = price * 15 / 100
else:
    discount = price * 10 / 100

net_amount = price - discount

print("Discount amount is:", discount)
print("Net amount to pay is:", net_amount)
```
