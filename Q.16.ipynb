```python
def calculate_power(base_value: int, power_value: int) -> int:
    """
    This funtcion calculate value of the power.
    """
    if not isinstance(base_value, int) or not isinstance(power_value,int):
        raise TypeError("Bothe base value and power value must be an integer")
    if power_value < 0:
        raise ValueError("Power value must be a possitive number")
    result = 1
    for i in range(power_value):
        result = result * base_value
    return result
if __name__ == "__main__":
    base_value = 2
    power_value = 3
    try:
        print(calculate_power(base_value,power_value))
    except (TypeError, ValueError) as error:
        print("Error", error)
    except Exception as e:
        print(e)
```
