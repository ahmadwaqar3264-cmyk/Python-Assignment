```python
n = 12
a = 0
b = 1
for i in range(n):
    print(a,end = " ")
    a,b = b, a+b
```

```python
def find_fibonacci(n: int) -> None:
    """
    This function find fibonacci.
    """
    if not isinstance(n,int):
        raise("n must be an integer")
    if n < 0:
        raise ValueError("n must be greater than zero")
    a = 0
    b = 1
    for i in range(n):
        print(a,end = " ")
        a,b = b, a+b  
if __name__ == "__main__":
    n = 12
    try:
        result = find_fibonacci(n)
    except (TypeError, ValueError) as e:
        print("Error", e)
    except Exception as e:
        print("Error", e)
```
