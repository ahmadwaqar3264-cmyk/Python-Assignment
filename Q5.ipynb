```python
num = 5
factorial = 1
for i in range(1, num+1):
    factorial = factorial * i
print(factorial)    
```

```python
def factorial_finder(num: int) -> int:
    """
    Calculate the factorial of non-negative integer
    """
    if isinstance(num, bool):
        raise TypeError("Boolean values are not allowed")
    if not isinstance(num, int):
        raise TypeError("Input must be an integer")
    if num < 0:
        raise ValueError("Factoial is not defined for negative numbers")
        
    factorial = 1
    for i in range(1, num+1):
        factorial = factorial * i
    return factorial
    
if __name__ == "__main__":
    try:
        number = input("Enter your number to find factorail : ")
        number = int(number)
        print("factorial", factorial_finder(number))
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print("error", e)
        
```
